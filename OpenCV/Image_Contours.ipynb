##**What are Contours ???** (boundaries of object)

Contours can be explained simply as a curve joining all the continuous points (along the boundary), having same color or intensity. The contours are a useful tool for shape analysis and object detection and recognition.

Note- binary images required



```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
im = cv2.imread('/content/objects.png')
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret,img_th = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)
contours, hierarchy = cv2.findContours(img_th,cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
#2nd-param - retrieval mode,3rd param - Contour approximation method
```

## Parameter - retrieval mode


---


RetrievalModes is a Hierarchy which shows how various contours are linked to each other, their relation with other contours, Parent Child relation


1.  cv.RETR_TREE

[next,prev,child,parent]

2.  cv.RETR_EXTERNAL

No Parent ,Child relation ship is given. All External contours are at same hierarchy (Level)

3. cv.RETR_LIST

It is similar to Tree but it does not establish any parent child relationship .
Retrieves all of the contours without establishing any hierarchical relationships.

4. cv.RETR_CCOMP

Retrieves all of the contours and organizes them into a two-level hierarchy.

##  Contour approximation method - Parameter

---



**cv.CHAIN_APPROX_NONE** - all the boundary points are stored

**cv.CHAIN_APPROX_SIMPLE** - It removes all redundant points and compresses the contour, thereby saving memory.
