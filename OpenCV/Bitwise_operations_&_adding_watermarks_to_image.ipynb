# Bitwise operations on images

The ! (bitwise NOT) :  0 becomes 1, and 1 becomes 0

The & (bitwise AND)  :  The result of AND is 1 only if both bits are 1.

The | (bitwise OR)  :  The result of OR is 1 if any of the two bits is 1.

The ^ (bitwise XOR)  :  The result of XOR is 1 if the two bits are different.



## 1. Bitwise NOT


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img1 = cv2.imread('/content/six.g',cv2.IMREAD_GRAYSCALE)
img1 = cv2.resize(img1, (500, 500)) 
cv2_imshow(img1)
```

```python
img1_not = cv2.bitwise_not(img1)
cv2.imwrite('img_out_not.jpeg',img1_not)
cv2_imshow(img1_not)
```

## 2. Bitwise AND 


```python
img_2= cv2.imread('/content/two.png',cv2.IMREAD_GRAYSCALE)
img_2 = cv2.resize(img_2, (500, 500)) 
img_6= cv2.imread('/content/six.png',cv2.IMREAD_GRAYSCALE)
img_6 = cv2.resize(img_6, (500, 500)) 
img_and=cv2.bitwise_and(img_6,img_2)
cv2_imshow(img_and)
```

## 3. Bitwise OR

```python
img_2= cv2.imread('/content/two.png',cv2.IMREAD_GRAYSCALE)
img_2 = cv2.resize(img_2, (500, 500)) 
img_6= cv2.imread('/content/six.png',cv2.IMREAD_GRAYSCALE)
img_6 = cv2.resize(img_6, (500, 500)) 
img_and=cv2.bitwise_or(img_6,img_2)
cv2_imshow(img_and)
```

## 2. Bitwise XOR

```python
img_2= cv2.imread('/content/two.png',cv2.IMREAD_GRAYSCALE)
img_2 = cv2.resize(img_2, (500, 500)) 
img_6= cv2.imread('/content/six.png',cv2.IMREAD_GRAYSCALE)
img_6 = cv2.resize(img_6, (500, 500)) 
img_and=cv2.bitwise_xor(img_6,img_2)
cv2_imshow(img_and)
```

# Adding watermark in images



```python
logo = cv2.imread("/content/dog_logo.png")
h_logo, w_logo, _ = logo.shape
```

```python
img = cv2.imread("/content/pet.png")
h_img, w_img, _ = img.shape
```

```python
center_y = int(h_img / 2)
center_x = int(w_img / 2)
top_y = center_y - int(h_logo / 2)
left_x = center_x - int(w_logo / 2)
bottom_y = top_y + h_logo
right_x = left_x + w_logo
```

```python
# Get ROI
roi = img[top_y: bottom_y, left_x: right_x]
# Image blendind ( adding transparency )
result = cv2.addWeighted(roi,1, logo, 0.1, 0.1)
# Replace the ROI on the image
img[top_y: bottom_y, left_x: right_x] = result
```

## Image masking 

```python
import cv2
import numpy as np

# Load image, create mask, and draw white circle on mask
image = cv2.imread('/content/flower.png')
mask = np.zeros(image.shape, dtype=np.uint8)
mask = cv2.circle(mask, (260, 300), 225, (255,255,255), -1) 

# Mask input image with binary mask
result = cv2.bitwise_and(image, mask)
# Color background white
```
