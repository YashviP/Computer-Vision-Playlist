# Geometric transformations of images (augmentation)

- Scaling 
- Translation
- Rotation
- Affine Transformation
- Perspective Transformation

````python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
````

## 1.  Scaling  - resizing image

````python
cv2_imshow(img)
````

```cv2.INTER_AREA``` - for shrinking 

```cv2.INTER_CUBIC (slow) & cv2.INTER_LINEAR (Default) ```  for zooming

````python
img = cv2.imread('/content/logo_2.png')

height, width = img.shape[:2]
res_z = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_LINEAR)
cv2_imshow(res_z)
````

## 2. Translation
- shifting of object location



image.png

````python
img = cv2.imread('/content/logo_2.png')
rows,cols,_ = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

cv2_imshow(dst)
````

## 3. Rotation


````python
img = cv2.imread('/content/logo_2.png')
rows,cols,_ = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))
````

````python
cv2_imshow(dst)
````

## 4. Affine Transformation

````python
import matplotlib.pyplot as plt
img = cv2.imread('/content/chess.png')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
````

## 5. Perspective Transformation

````python
img = cv2.imread('/content/chess_2.png')
rows,cols,ch = img.shape

pts1 = np.float32([[75,175],[175,175],[175,75],[75,75]])
pts2 = np.float32([[0,250],[250,250],[250,0],[0,0]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
````
