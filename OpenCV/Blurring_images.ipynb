# *What is Filter ??*

A filter is defined by a kernel, which is a small array applied to each pixel and its neighbors within an image


###  *Low Pass Filter*

low pass filter tends to retain the low frequency information within an image while reducing the high frequency information. ( smoothing or blurring )

### *High Pass Filter*

A high pass filter tends to retain the high frequency information within an image while reducing the low frequency information. ( sharpening )

# Blurring 

Blurring is done by performing convolution operation between the given image and a low-pass filter kernel. 



### *Types of filters for blurring -*

1. Averaging
2. Gaussian Filtering
3. Bilateral Filtering
4. Median Filtering

```python
# Import the image
img = cv2.imread('/content/trees.png', 1)

def average_kernel(size):
  k = np.ones((size,size),np.float32)/(size**2)
  return k

size=5
avg_filter_img = cv2.filter2D(img,-1,average_kernel(size))
cv2_imshow(avg_filter_img)

# Or 

blur = cv2.blur(img,(5,5))
```

# Bilateral Filter

Bilateral filtering also takes a Gaussian filter in space, but additionally considers one more Gaussian filter which is a function of pixel difference. 

The Gaussian function of space makes sure that only nearby pixels are considered for blurring, while the Gaussian function of intensity difference makes sure that only those pixels with similar intensities to the central pixel are considered for blurring. 

```python
import cv2
```

```python
img = cv2.imread('/content/bl.jpg',1)
```

```python
import matplotlib.pyplot as plt
def plot_img(images, titles):
  fig, axs = plt.subplots(nrows = 1, ncols = len(images),
                          figsize = (20, 20))
  for i, p in enumerate(images):
    axs[i].imshow(cv2.cvtColor(p, cv2.COLOR_BGR2RGB))
    axs[i].set_title(titles[i])
  plt.show()


g_img = cv2.GaussianBlur(img,(7,7),0)
b_img = cv2.bilateralFilter(img,7,75,75) 
# d - diameter of each pixel neighbourhood
# sigma s - for gaussian filter in space
# sigma r - for intesity difference 
images=[img, g_img, b_img]
titles=['original image',
        'gaussian filter image',
        'Bilateral filter image']
plot_img(images, titles)
```
