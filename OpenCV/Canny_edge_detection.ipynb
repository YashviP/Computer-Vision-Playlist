The Canny edge detection is a popular edge detection algorithm uses a multi-stage algorithm to detect a wide range of edges in images.


   
**The Canny edge detection algorithm is composed of 4 steps:**

* Noise reduction
* Gradient calculation
* Non-maximum suppression
* Edge Tracking by Hysteresis

## Step 3 - Non-Maximum Suppression

Non maximum suppression works by finding the pixel with the maximum value in an edge. 

image.png

* Given the current pixel, find two neighboring pixels in the same direction as the computed orientation of current pixel.

* Compare the magnitudes from three (current and two neighboring pixels)
If the magnitude of current pixel is the maximum, keep its magnitude. Otherwise discard the pixel from the edge detection process.

## Step 4 : Hysteresis Thresholding


***Apply thresholding***

If the value of any pixel is higher than the **high value**, then we set it to 255. We assume these are proper edges. Next if the pixels are between **low** and **high** value then we set them to weak value(say 50 )

**Apply hysteresis**

whether any given pixels neighbors has value equal to 255, if yes then change the value of the pixel to 255, otherwise discard the pixel by setting the value to 0.

```python
img = cv2.imread('/content/building.jpeg',0)
edges = cv2.Canny(img,100,255,3,L2gradient = True) # if l2_gradient==False(default) then Edge gradient = |Gx| + |Gy|
images=[img,edges]
titles=['original image',
        'edge image']
plot_img(images, titles)
```
