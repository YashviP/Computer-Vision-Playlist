## Draw shapes in an Image 
* Rectangle 
* Line
* circle 
* Polygon
* PutText

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

## Drawing line

```python
# Create a black image
img = np.zeros((512,512,3), np.uint8)

# Draw a diagonal line with thickness of 3 px
img = cv2.line(img,(50,100),(455,431),(255,0,0),3)
```

```python
cv2_imshow(img)
```

## Drawing rectangle 

```python
img = cv2.rectangle(img,(400,0),(510,140),(0,255,0),3)
```

```python
img_2 = cv2.rectangle(img,(400,0),(510,140),(0,255,0),-1)
```

```python
cv2_imshow(img_2)
```

## Drawing polygon

```python
pts = np.array([[300,500],[200,300],[400,200],[450,140]], np.int32)
pts = pts.reshape((-1,1,2))
img = cv2.polylines(img,[pts],True,(0,255,255))
cv2_imshow(img)
```

## Drawing circle 

```python
img = cv2.circle(img,(400,400), 52, (0,0,255),3)
```

```python
cv2_imshow(img)
```

```python
img = cv2.circle(img,(300,200),50, (0,0,255),3)
cv2_imshow(img)
```

## Adding text

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'YASHVI',(200,450), font,3,(255,255,255),5,cv2.LINE_AA)
```
