# Introduction to OpenCV
OpenCV(Open Source Computer Vision). It is a open source computer vision library that allows you to manipulate images and videos. OpenCV includes a comprehensive set of both classic and state-of-the-art computer vision and machine learning algorithms. These algorithms can be used to detect and recognize faces, identify objects, classify human actions in videos, track camera movements, track moving object etc etc.

# Getting started with Images 

``` imread ```  -  load an image 

``` imwrite ``` -  write to image

``` imshow ```  -  display image 

image => Multidimensional array of pixels ( 0-255 )

````
import numpy as np
import cv2
````

````
# lets create one black image 
black_img=np.zeros((5,5),dtype=np.uint8)
````

````
black_img
````

````
black_img_bgr = cv2.cvtColor(black_img, cv2.COLOR_GRAY2BGR)
black_img_bgr
````

### Read Image 

````
image=cv2.imread('/content/Take_in_the_Scenery.jpg')
````

```IMREAD_COLOR``` loads the image in the BGR 8-bit format. This is the default that is used here.

```IMREAD_UNCHANGED``` loads the image as is (including the alpha channel if present)

```IMREAD_GRAYSCALE``` loads the image as an intensity one

````
image_grayscale=cv2.imread('/content/Take_in_the_Scenery.jpg',cv2.IMREAD_GRAYSCALE)
````

````
image_alpha=cv2.imread('/content/car.png',cv2.IMREAD_UNCHANGED)
image_without_alpha=cv2.imread('/content/car.png')
````

````
image_grayscale
````

````
image
````

### Write to Image

````
cv2.imwrite('out.jpg',image)
````
